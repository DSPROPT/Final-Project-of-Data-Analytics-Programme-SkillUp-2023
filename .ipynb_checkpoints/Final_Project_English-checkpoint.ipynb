## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`

### Brookln:(This seems to be a typo or variant spelling of Brooklyn)
### Manhatan:  (This seems to be a typo or variant spelling of Manhattan)



```python
# Define a function to correct the typos
def correct_typos(value):
    if pd.isna(value):
        return value
    corrections = {
        'brookln': 'Brooklyn',
        'manhatan': 'Manhattan'
    }
    return corrections.get(value.lower(), value)

# Apply the function to the 'neighbourhood_group' column
df_cleaned['neighbourhood_group'] = df_cleaned['neighbourhood_group'].apply(correct_typos)

df_cleaned['neighbourhood_group'].unique()
```

```python
## Which room type adheres to more strict cancellation policy
# Room type with the most strict cancellation policy
strict_cancellation = df_cleaned[df_cleaned['cancellation_policy'] == 'strict']
strict_cancellation_count = strict_cancellation['room_type'].value_counts().idxmax()

print('Room type with the most strict cancellation policy is',strict_cancellation_count)
```

```python
## List the prices by neighborhood group and also mention which is the most expensive neighborhood group for rentals
average_price_neighborhood = df_cleaned.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
most_expensive_neighborhood_group = average_price_neighborhood.idxmax()
most_expensive_neighborhood_group_average_price = average_price_neighborhood.max()



print(average_price_neighborhood)
print('the most expensive neighborhood group for rentals is',most_expensive_neighborhood_group) 


print('with the average_price of',most_expensive_neighborhood_group_average_price,'$')
```

* List the count of various room types avaliable with Airnb

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Setting up the style and size for the plots
sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))

# 1. List the count of various room types available with Airbnb
plt.subplot(1, 2, 1)
sns.countplot(y='room_type', data=df_cleaned, order = df_cleaned['room_type'].value_counts().index)
plt.title('Count of Various Room Types')
plt.xlabel('Count')
plt.ylabel('Room Type')

# 2. Which room type adheres to more strict cancellation policy
plt.subplot(1, 2, 2)
strict_data = df_cleaned[df_cleaned['cancellation_policy'] == 'strict']
sns.countplot(y='room_type', data=strict_data, order = strict_data['room_type'].value_counts().index)
plt.title('Room Types with Strict Cancellation Policy')
plt.xlabel('Count')
plt.ylabel('Room Type')

plt.tight_layout()
plt.show()
```

```python
df_cleaned['neighbourhood_group'].dtype
```

```python
df_cleaned['price'].dtype
```

- The first two visualizations have been generated:

- Count of Various Room Types with Airbnb:

- The most common room type is "Entire home/apt", followed by "Private room". "Shared room" and "Hotel room" are considerably less common.
Room Types with Strict Cancellation Policy:

- Among the listings with a strict cancellation policy, "Entire home/apt" is the most common room type.

```python
# 3. List the prices by neighborhood group and mention which is the most expensive neighborhood group for rentals

plt.figure(figsize=(14, 6))
# Sort the dataframe by the mean price for better visualization
sorted_data = df_cleaned.groupby('neighbourhood_group')['price'].mean().sort_values().reset_index()
# Plot using the sorted data
sns.barplot(x='neighbourhood_group', y='price', data=sorted_data, estimator=np.nanmean, palette='viridis', errorbar=None)
plt.title('Average Price by Neighbourhood Group', fontsize=16)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.xticks(rotation=45)
# Add value labels on top of the bars
for index, value in enumerate(sorted_data['price']):
    plt.text(index, value, f"{value:.2f}", ha='center', va='bottom', fontsize=10)
# Add horizontal gridlines for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Despine for a cleaner look
sns.despine()
plt.show()
```

### Average Price by Neighbourhood Group:
- The neighborhood group "Queens" appears to have the highest average price, making it the most expensive neighborhood group for rentals.

```python
# 4. List the top 10 neighborhoods in increasing order of their price with a horizontal bar graph
top_neighbourhoods = df_cleaned.groupby('neighbourhood')['price'].mean().sort_values().head(20)

plt.figure(figsize=(14, 7))
top_neighbourhoods.plot(kind='barh', color=sns.color_palette('viridis', 10))
plt.title('Top 10 Neighbourhoods with Lowest Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.show()
```

### Top 10 Neighbourhoods with Lowest Average Price:
- The horizontal bar graph displays the top 10 neighborhoods with the lowest average prices. **"Lighthouse Hill"** is the neighborhood with the lowest average price, making it the cheapest neighborhood based on the data provided.

```python
# 5. List the neighborhoods which offer short term rentals within 10 days
short_term_neighbourhoods = df_cleaned[df_cleaned['minimum_nights'] <= 10].groupby('neighbourhood')['neighbourhood'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
short_term_neighbourhoods.plot(kind='bar', color=sns.color_palette('muted'))
plt.title('Top 10 Neighbourhoods Offering Short Term Rentals (â‰¤ 10 days)')
plt.ylabel('Number of Listings')
plt.xlabel('Neighbourhood')
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Neighbourhoods Offering Short Term Rentals (â‰¤ 10 days):
- The bar graph displays the top 10 neighborhoods with the most listings offering short-term rentals within 10 days. **"Bedford-Stuyvesant"** is the neighborhood with the highest number of such listings.

```python
# 6. List the prices with respect to room type using a bar graph
# Calculate average prices
average_prices = df_cleaned.groupby('room_type')['price'].mean().sort_values(ascending=False)
sorted_room_types = average_prices.index

# Create the plot
plt.figure(figsize=(14, 7))
barplot = sns.barplot(x='room_type', y='price', data=df_cleaned, order=sorted_room_types,
                      estimator=np.nanmean, errorbar=None, palette='Spectral')

# Annotate each bar with the average price
for i, room_type in enumerate(sorted_room_types):
    price = average_prices[room_type]
    plt.text(i, price, f'{price:.2f}', ha='center', va='bottom')

# Enhance aesthetics
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
sns.despine()

# Rotate x labels if needed
plt.xticks(rotation=45)
plt.show()
```

### Average Price by Room Type:
- The bar graph displays the average price for each room type. **"Hotel room" has the highest average price, while "Private room" has the lowest average price.**


```python
# 7. Create a pie chart that shows distribution of booked days for each neighborhood group
booked_days_neighbourhood = df_cleaned.groupby('neighbourhood_group')['days_booked'].sum()

# Calculate the percentage values
percentages = 100 * booked_days_neighbourhood / booked_days_neighbourhood.sum()

# Create custom labels for the legend
custom_labels = [f"{index} ({percentage:.1f}%)" for index, percentage in percentages.items()]

# Increase the figure size and plot the pie chart with improved visuals
plt.figure(figsize=(16, 12))
booked_days_neighbourhood.plot(kind='pie', labels=None, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'), wedgeprops=dict(width=0.3), pctdistance=0.85)

# Draw a circle in the center to make it a donut chart (this can help improve visibility for smaller values)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribution of Booked Days by Neighbourhood Group')
plt.ylabel('')  # Hide the default y-label which is 'days_booked'

# Add the custom labels to the legend
plt.legend(custom_labels, loc='upper left', bbox_to_anchor=(1, 0.5))

plt.show()
```

### Distribution of Booked Days by Neighbourhood Group:
- The pie chart displays the distribution of booked days for each neighborhood group.
- **"Manhattan" occupies the largest portion**, indicating that it has the highest booking percentage among all the neighborhood groups. This suggests that Manhattan is the most popular neighborhood group for Airbnb bookings based on the data provided.

## Task 5c: Data Visualization (Any Tool)

* With the help of box plots illustrate the following
 * Effect of Review Rate number on price
 * Effect of host identity verified on price

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# Convert 'review_rate_number' to a categorical variable
df_cleaned['review_rate_number'] = df_cleaned['review_rate_number'].astype('category')

# Then, make sure to handle NaN values in 'host_identity_verified'
# For the purpose of visualization, you might want to drop these or replace with a category like 'unknown'
df_cleaned['host_identity_verified'] = df_cleaned['host_identity_verified'].fillna('unknown')
df_cleaned['host_identity_verified'] = df_cleaned['host_identity_verified'].astype('category')

# Now, let's create the boxplots
plt.figure(figsize=(16, 7))

# 1. Effect of Review Rate number on price
plt.subplot(1, 2, 1)
sns.boxplot(x='review_rate_number', y='price', data=df_cleaned, palette='viridis')
plt.title('Effect of Review Rate Number on Price')
plt.xlabel('Review Rate Number')
plt.ylabel('Price ($)')

# 2. Effect of host identity verified on price
plt.subplot(1, 2, 2)
sns.boxplot(x='host_identity_verified', y='price', data=df_cleaned, palette='pastel')
plt.title('Effect of Host Identity Verified on Price')
plt.xlabel('Host Identity Verified')
plt.ylabel('Price ($)')

plt.tight_layout()
plt.show()
```

## The two box plots have been generated to illustrate:

#### Review Rate Number on Price:

* The boxplots across different review rates from 1 to 5 show overlapping interquartile ranges, indicating that there's not a clear distinction in price based on the review rate number.
* There's a consistency in median prices across all review rate numbers, suggesting that review rate number may not be a strong predictor of price.
* There are no visible outliers in the data, which could mean that outlier values have been removed or that the dataset lacks extreme values in price.

#### Host Identity Verified on Price:

* The three categories, 'unconfirmed', 'unknown', and 'verified', show some variation in median prices, with 'unconfirmed' and 'verified' having higher median prices than 'unknown'.
* The spread of prices (as indicated by the interquartile range) is similar across all three categories, although 'verified' shows a slightly larger range, suggesting more variability in price for listings with verified hosts.
* Similar to the review rate number boxplots, there are no outliers shown here.

### Conclusions:

* The review rate number doesn't seem to have a significant effect on the listing price, as all review rate categories have similar price distributions.
* Listings with verified hosts tend to have a wider range of prices, potentially indicating that verified listings might offer more premium options, which could explain the higher variability.
* The 'unknown' category for host identity verification has a lower median price, which might suggest that guests could be paying less for listings where the host's identity status is not confirmed or provided.
