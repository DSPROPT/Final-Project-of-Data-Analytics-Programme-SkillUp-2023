Removing these columns can make the dataset more concise without losing meaningful information. If the dataset included listings from multiple countries, these columns would be more relevant.



## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
missing_values_count = df_cleaned.isnull().sum().sort_values()

# Impute missing values based on datatype of the columns
for column in df_cleaned.columns:
    if df_cleaned[column].dtype == 'object':
        df_cleaned[column].fillna(df_cleaned[column].mode()[0], inplace=True)
    else:
        df_cleaned[column].fillna(df_cleaned[column].mean(), inplace=True)
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
# Check for duplicate values in the dataframe and remove them
initial_records_count = len(df_cleaned)
df_cleaned.drop_duplicates(inplace=True)
after_deduplication_records_count = len(df_cleaned)
```

```python
## Display the total number of records in the dataframe after removing the duplicates.
missing_values_count, initial_records_count, after_deduplication_records_count
```

## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# Rename the column 'availability 365' to 'days_booked'
df_cleaned.rename(columns={'availability 365': 'days_booked'}, inplace=True)
```

```python
# Convert all column names to lowercase and replace spaces with underscores
df_cleaned.columns = df_cleaned.columns.str.lower().str.replace(' ', '_')
```

```python
# Remove dollar sign and comma from 'price' and 'service_fee' columns and convert to float data type
df_cleaned['price'] = df_cleaned['price'].str.replace('[,$]', '', regex=True).astype(float)
df_cleaned['service_fee'] = df_cleaned['service_fee'].str.replace('[,$]', '', regex=True).astype(float)


# Display the first five rows of the transformed dataframe to verify the changes
df_transformed_head = df_cleaned.head()

df_transformed_head
```

### The transformations have been successfully applied to the dataframe:

- The column availability 365 was renamed to days_booked.
- All column names were converted to lowercase, and spaces were replaced with underscores.
- Dollar signs and commas were removed from the price and service_fee columns, and these columns were converted to the float data type.
- The first five rows of the transformed dataframe are displayed above for verification.

## Task 5a: Data Visualization (Any Tool)

* List the count of various room types avaliable with Airnb
* Which room type adheres to more strict cancellation policy
* List the prices by neighborhood group and also mention which is the most expensive neighborhood group for rentals
* List the top 10 neighborhoods in the increasing order of their price with the help of a horizontal bar graph. Which is the cheapest neighborhood.
* List the neighborhoods which offer short term rentals within 10 days. Illustrate with a bar graph
* List the prices with respect to room type using a bar graph and also state your inferences.
* Create a pie chart that shows  distribution of booked days for each neighborhood group .Which neighborhood has the highest booking percentage.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the style and size for the plots
sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))

# 1. List the count of various room types available with Airbnb
plt.subplot(1, 2, 1)
sns.countplot(y='room_type', data=df_cleaned, order = df_cleaned['room_type'].value_counts().index)
plt.title('Count of Various Room Types')
plt.xlabel('Count')
plt.ylabel('Room Type')

# 2. Which room type adheres to more strict cancellation policy
plt.subplot(1, 2, 2)
strict_data = df_cleaned[df_cleaned['cancellation_policy'] == 'strict']
sns.countplot(y='room_type', data=strict_data, order = strict_data['room_type'].value_counts().index)
plt.title('Room Types with Strict Cancellation Policy')
plt.xlabel('Count')
plt.ylabel('Room Type')

plt.tight_layout()
plt.show()
```

- The first two visualizations have been generated:

- Count of Various Room Types with Airbnb:

- The most common room type is "Entire home/apt", followed by "Private room". "Shared room" and "Hotel room" are considerably less common.
Room Types with Strict Cancellation Policy:

- Among the listings with a strict cancellation policy, "Entire home/apt" is the most common room type.

```python
# 3. List the prices by neighborhood group and mention which is the most expensive neighborhood group for rentals
plt.figure(figsize=(14, 6))
sns.barplot(x='neighbourhood_group', y='price', data=df_cleaned, estimator=lambda x: sum(x)/len(x), errorbar=None)
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

### Average Price by Neighbourhood Group:
- The neighborhood group "Queens" appears to have the highest average price, making it the most expensive neighborhood group for rentals. "Manhatan" (possibly a typo for Manhattan) has the lowest average price among the groups.

```python
# 4. List the top 10 neighborhoods in increasing order of their price with a horizontal bar graph
top_neighbourhoods = df_cleaned.groupby('neighbourhood')['price'].mean().sort_values().head(10)

plt.figure(figsize=(14, 7))
top_neighbourhoods.plot(kind='barh', color=sns.color_palette('viridis', 10))
plt.title('Top 10 Neighbourhoods with Lowest Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.show()
```

### Top 10 Neighbourhoods with Lowest Average Price:
- The horizontal bar graph displays the top 10 neighborhoods with the lowest average prices. "Bull's Head" is the neighborhood with the lowest average price, making it the cheapest neighborhood based on the data provided.

```python
# 5. List the neighborhoods which offer short term rentals within 10 days
short_term_neighbourhoods = df_cleaned[df_cleaned['minimum_nights'] <= 10].groupby('neighbourhood')['neighbourhood'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
short_term_neighbourhoods.plot(kind='bar', color=sns.color_palette('muted'))
plt.title('Top 10 Neighbourhoods Offering Short Term Rentals (â‰¤ 10 days)')
plt.ylabel('Number of Listings')
plt.xlabel('Neighbourhood')
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Neighbourhoods Offering Short Term Rentals (â‰¤ 10 days):
- The bar graph displays the top 10 neighborhoods with the most listings offering short-term rentals within 10 days. "Williamsburg" is the neighborhood with the highest number of such listings.

```python
# 6. List the prices with respect to room type using a bar graph
plt.figure(figsize=(14, 7))
sns.barplot(x='room_type', y='price', data=df_cleaned, estimator=lambda x: sum(x)/len(x), errorbar=None, palette='pastel')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.show()
```

### Average Price by Room Type:
- The bar graph displays the average price for each room type. "Entire home/apt" has the highest average price, while "Shared room" has the lowest average price.

## Inferences:

- As expected, listings that offer an entire home or apartment tend to be more expensive than those offering private rooms or shared spaces. This is likely due to the increased privacy, space, and amenities that entire homes or apartments provide.
- Shared rooms are the least expensive, which makes sense since guests are sharing the space with others and might have less privacy.
- Next, let's create a pie chart that shows the distribution of booked days for each neighborhood group.

```python
# 7. Create a pie chart that shows distribution of booked days for each neighborhood group
booked_days_neighbourhood = df_cleaned.groupby('neighbourhood_group')['days_booked'].sum()

# Calculate the percentage values
percentages = 100 * booked_days_neighbourhood / booked_days_neighbourhood.sum()

# Create custom labels for the legend
custom_labels = [f"{index} ({percentage:.1f}%)" for index, percentage in percentages.items()]

# Increase the figure size and plot the pie chart with improved visuals
plt.figure(figsize=(16, 12))
booked_days_neighbourhood.plot(kind='pie', labels=None, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'), wedgeprops=dict(width=0.3), pctdistance=0.85)

# Draw a circle in the center to make it a donut chart (this can help improve visibility for smaller values)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribution of Booked Days by Neighbourhood Group')
plt.ylabel('')  # Hide the default y-label which is 'days_booked'

# Add the custom labels to the legend
plt.legend(custom_labels, loc='upper left', bbox_to_anchor=(1, 0.5))

plt.show()
```

### Distribution of Booked Days by Neighbourhood Group:
- The pie chart displays the distribution of booked days for each neighborhood group.
- "Manhattan" occupies the largest portion, indicating that it has the highest booking percentage among all the neighborhood groups. This suggests that Manhattan is the most popular neighborhood group for Airbnb bookings based on the data provided.

- Next, let's create a line graph to show in which year the maximum construction of rooms took place.

```python
# 2. Using a line graph show in which year the maximum construction of rooms took place
# First, we will extract the year from the 'construction_year' column and then count the number of rooms constructed each year

# Extracting the year and counting the number of constructions per year
construction_counts = df_cleaned['construction_year'].value_counts().sort_index()

plt.figure(figsize=(14, 7))
construction_counts.plot(kind='line', marker='o')
plt.title('Number of Room Constructions by Year')
plt.xlabel('Year')
plt.ylabel('Number of Constructions')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

### Number of Room Constructions by Year:
- The line graph depicts the number of room constructions over the years.


### Inferences:

- There is a notable increase in constructions from the early 1900s, peaking around the mid-20th century.
- Post that peak, there's a decline with some fluctuation, and we observe a steady increase again in recent years.
- It appears that the maximum construction of rooms occurred around the mid-20th century.
